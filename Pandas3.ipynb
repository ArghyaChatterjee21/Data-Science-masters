```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]})
data1
```

```python
data2 = pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]})
data2
```

```python
## Inner = Common part by default

pd.merge(data1,data2)
```

```python
## Left part

pd.merge(data1,data2, how = 'left')
```

```python
## Right part

pd.merge(data1,data2, how = 'right')
```

```python
## Outer part

pd.merge(data1,data2, how = 'outer')
```

```python
data1 
```

```python
data2
```

```python
## Cross merge

pd.merge(data1,data2, how = 'cross')
```

```python
pd.merge(data1, data2 , how = 'left', on = 'key1')
```

```python
## Specific column name
pd.merge(data1, data2 , how = 'left',left_on = 'key2', right_on = 'key4')
```

```python
Data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]},
                     index = ['a','b','c','d','e'])
Data1
```

```python
Data2 = pd.DataFrame({'key11':[1,2,4,5,6],
                      'key22':[4,5,6,7,8],
                      'key33':[3,4,5,6,6]},
                     index=['a','b','h','i','j'])
Data2
```

```python
## Left join by default
## Row wise
Data1.join(Data2)
```

```python
## Right join 
## Row wise
Data1.join(Data2,how = 'right')
```

```python
## Common part
Data1.join(Data2,how = 'inner')
```

```python
Data1.join(Data2,how = 'outer')
```

```python
Data1.join(Data2,how='cross')
```

```python
## Create a new column where fare(euro) turns into fare(inr)

df["Fare_inr"] = df["Fare"].apply(lambda x : x*80)
```

```python
df.head(3)
```

```python
## Create a new column where length of name will available
df["Name_len"] = df["Name"].apply(len)
```

```python
df.head(3)
```

```python
## Without using lambda function
def inr_con(x):
    return x*80
df["Fare_inr"] = df["Fare"].apply(inr_con)
```

```python
df.head(3)
```

```python
## Divided into  3 categories acoording to their fare

def create_flag(x):
    if x< 10 :
        return "cheap"
    elif x >= 10 and x < 20 :
        return  "mid"
    else :
        return "high"
```

```python
df["Flag_fare"] = df["Fare"].apply(create_flag)
```
