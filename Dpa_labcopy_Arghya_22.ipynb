1. Write a python program to create two dictionaries and print a third dictionary by taking the union of their keys.

```python
d1={1:20 ,2:25, 3:65, 6:50}
d2={1:78 ,2:14, 4:88, 5:13}
newd=d2
for i,j in d1.items():
    if i in d2:
        newd[i]=j+d2[i]
    else:
        newd[i]=j
print('New Dictionary : ',newd)
```

```python
sortd=list(newd.keys())
sortd.sort()
sortdict={i : newd[i] for i in sortd}
sortdict            #sorted by key values
```

3. Write a python program to create .csv file in your system. Write the file with the csv package. Read files delimited by tab, colon, semicolon or any other standard delimiters.

```python
import csv
import re
# create a CSV file with some data
filename = "example.csv"
with open(filename, 'w', newline='') as file:
    writer = csv.writer(file, delimiter=',')
    writer.writerow(["Name", "Age", "City"])
    writer.writerow(["Aryan", 25, "Singapore"])
    writer.writerow(["Binod", 30, "Paris"])
    writer.writerow(["Lizanki", 35, "London"])
# read the file with different delimiters
delimiters = [',', '\t', ':', ';']  # list of possible delimiters
for delimiter in delimiters:
    with open(filename, 'r') as file:
        reader = csv.reader(file, delimiter=delimiter)
        print(f"Reading file with delimiter '{delimiter}':")
        for row in reader:
            text = ','.join([i for i in row])
            #print(text)
            print(re.sub("[,]", f"'{delimiter}'", text))
        print()  # add a blank line between outputs
```

10. Write a Python Program to write a csv file using for loop to store the marks a student had obtained in the B Tech semesters. Ask user for the input.

```python
import csv
filename=("result_btech.csv")
with open(filename, 'w', newline='') as file:
    writer = csv.writer(file, delimiter=',')
    writer.writerow(["Semester", "Subject-1", "Subject-2", "Subject-3", "Subject-4"])
    x=int(input("Write the number of semesters : "))
    for i in range(0,x):
        print("Enter marks for semester",i+1)
        s1=input("Enter marks for subject-1 : ")
        s2=input("Enter marks for subject-2 : ")
        s3=input("Enter marks for subject-3 : ")
        s4=input("Enter marks for subject-4 : ")
        writer.writerow([i,i+1, s1, s2, s3, s4])
```

13. Write a Python program to find Mean, Median and Mode.

```python
n=int(input("Enter Total Number of Elements : "))
l=[]
sum=0
for i in range(0,n):
    x=int(input('Enter element : '))
    l.append(x)
    sum=sum+x
mn=sum/n
l1=l
print("Mean of the given numbers (",l1,")   : ",mn)

no = len(l)
l.sort()
if no % 2 == 0:
    median1 = l[no//2]
    median2 = l[no//2 - 1]
    median = (median1 + median2)/2
else:
    median = l[no//2]
print("Median of the given numbers (",l1,") : ", median)

def mode1(l):
    y={}
    for a in l:
        if not a in y:
            y[a]=1
        else:
            y[a]+=1
    return [g for g,l in y.items() if l==max(y.values())]

print("Mode of the given numbers (",l1,")   : ",mode1(l))
```
