## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize = (6,3))   ## To control graph size
plt.scatter(x, y, c = "blue", alpha = .9, marker = "x")  ## C= Color Alpha = Color intensity Marker = Shape wheather it is dot or cross
plt.xlabel("X-Axis")   ## Naming x axis
plt.ylabel("Y-Axis")   ## Naming y axis
plt.title("X Vs Y")    ## Giving title
plt.grid()             ## To make grid
```

```python
x1 = np.linspace(1, 10, 100) ## Return evenly spaced numbers over a specified interval.
y1 = np.sin(x1)              ## The sine of each element of x1  
```

```python
x1
```

```python
y1
```

```python
plt.plot(x1,y1, c = 'green')
```

```python
plt.scatter(x1,y1)
```

```python
plt.figure(figsize = (4,4))
plt.plot(x1,y1, "--g")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("sin wave graph")
plt.grid()
```

```python
x2 = ['a','b','c','d','e']
y2 = np.random.rand(5)
```

```python
x2
```

```python
y2
```

```python
plt.figure(figsize = (5,3))
plt.bar(x2,y2, color = 'cyan')
plt.xlabel("Cat data")
plt.ylabel("Num data")
plt.title("Bar Plot")
plt.grid()
```

```python
plt.figure(figsize = (5,3))
plt.barh(x2,y2, color = 'cyan')    ## To change x and y axis 
plt.xlabel("Num data")
plt.ylabel("Cat data")
plt.title("Bar Plot")
plt.grid()
```

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
```

```python
plt.figure(figsize = (6,3))
plt.hist(data, color = "#700EBE") ## y axis represents the frequwency of the data
plt.xlabel("Unique Val")
plt.ylabel("Frequency")
plt.title("Hist Plot")
plt.grid()
plt.savefig("Hist.jpg") ## To save the picture
```

#### 3D Graph

```python
X = np.random.rand(50)
Y = np.random.rand(50)
Z = np.random.rand(50)

## Create a new figure, or activate an existing figure
fig = plt.figure() 

## The add_subplot() method figure module of matplotlib library is used to add an Axes to the figure as part of a subplot arrangement.
ax = fig.add_subplot(projection = '3d')
ax.scatter(X, Y, Z, c ='r', alpha = .8)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```
